```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.linspace(0, 2*np.pi, 1000)
f = x
g = np.sin(x)

plt.plot(x, f, label="f=x", markersize=3)
plt.plot(x, g, label="g=sin(x)", markersize=3)

# Add legend, title and show plots
plt.legend()
plt.title("All graphs together")
plt.show()
```
